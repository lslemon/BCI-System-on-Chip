```python
def plot_to_notebook(time_sec, input_sig, title, samples = None, out_signal=None):
    print('Number of samples: ',len(input_sig))
    plt.rcParams["figure.figsize"] = (20,10)
    plt.rcParams["font.size"] = 22
    plt.figure()
    plt.subplot(1,1,1)
    plt.xlabel("Time (seconds)")
    plt.ylabel("Voltage (uV)")
    plt.grid()
    if samples is None:
        plt.plot(time_sec, input_sig, 'y-', label=title)
    else:    
        plt.plot(time_sec[:samples], input_sig[:samples], 'y-', label=title)

    if out_signal is not None:
        plt.plot(time_sec, out_signal, 'g-', label='Filtered Signal')
    
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
def plot_fir_response(coeffs, fs):
    """Plot the frequency magnitude response of a set of FIR filter weights"""
    
    freqs, resp = freqz(coeffs, 1)
    plt.plot(freqs/math.pi, np.abs(resp))
    plt.rcParams["figure.figsize"] = (20,10)
    plt.show()
```

```python
def filter_signal(signal, f_samp, low_cut, high_cut, taps):
#     nyq = f_samp / 2.0
#     # Design high-pass filter with cut-off at 2.8 kHz
#     hpf_coeffs = firwin(taps, [low_cut/nyq, high_cut/nyq], pass_zero=False)
    
    hpf_coeffs = filter_design(f_samp, low_cut, high_cut, taps)
    
    # Plot our filter's frequency response as a sanity check
    plot_fir_response(hpf_coeffs, f_samp)
    filtered_signal = lfilter(hpf_coeffs, 1.0, signal) 
    
    np.save("hpf_coeffs.npy", hpf_coeffs)
    
    return hpf_coeffs, filtered_signal
```

```python
def filter_design(f_samp, low_cut, high_cut, taps):
    nyq = f_samp / 2.0
    # Design high-pass filter with cut-off at 2.8 kHz
    coeffs = firwin(taps, [low_cut/nyq, high_cut/nyq], pass_zero=False)
    return coeffs
```

```python
def to_freq_dataframe(samples, fs):
    """Create a pandas dataframe from an ndarray frequency domain samples"""
    sample_freqs = np.linspace(0, fs, len(samples))
    return pd.DataFrame(dict(
        amplitude = samples[0:int(len(samples)/2)],  
        freq      = sample_freqs[0:int(len(samples)/2)]
    ))

def plot_fft_spectrum(fft_complex, f_samp, channel_name):
    fft_magnitude = np.abs(fft_complex)
    dataframe = to_freq_dataframe(20*np.log10(fft_magnitude), f_samp)
    freqs = dataframe["freq"]
    amps = dataframe["amplitude"]
    
    max_amps = plt.ylim()
    print(max_amps)
    plt.rcParams["figure.figsize"] = (20,10)
    plt.plot(freqs, amps)
    
#     plt.axvline(10,color='r', ymax = 20/max_amps[1])
#     plt.stem([10], [20])
#     plt.axhline(57)
#     plt.axvline(8.57,color='r')
#     plt.axvline(7.5,color='r')
#     plt.axvline(6.66,color='r')
    
#     plt.axvline(24, color='g')
#     plt.axvline(20, color='g')
#     plt.axvline(17.14, color='g')
#     plt.axvline(15, color='g')
#     plt.axvline(13.32, color='g')
    plt.title(channel_name)

    
    
    plt.grid()
    
    plt.show()
    
    return dataframe
```
